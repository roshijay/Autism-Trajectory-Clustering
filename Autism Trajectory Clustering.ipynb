# Modeling Developmental Heterogeneity in Autism via Longitudinal Unsupervisd Clustering

**Abstract**

Autism Spectrum Disorder (ASD) is a neurodevelopmental condition marked by significant heterogeneity in cognitive, social, communicative, and behavioral functioning. This phenotypic variability presents challenges for generalizing developmental trajectories and tailoring effective interventions. Historically, research has relied heavily on cross-sectional study designs, which capture static snapshots in time and fail to account for the temporal dynamics of autistic development.

This study applies longitudinal clustering, an unsupervised machine learning technique, to identify trajectory-based subgroups in cognitive development across multiple childhood and adolescent time points. Using a synthetic dataset of 200 individuals, demographically inclusive and ethically constructed to reflect real-world diversity, we modeled developmental change across five ages (3, 6, 9, 12, and 15 years). Each individual was assigned to one of three latent trajectory types: improving, stable, or declining.

Our results show that longitudinal clustering can successfully differentiate between these subtypes, capturing both the rate and direction of developmental change. Models at k = 3 and k = 6 provided especially rich insight, with Adjusted Rand Indices indicating moderate alignment with ground-truth labels despite measurement noise. Importantly, clustering outcomes were not driven by gender or ethnicity, supporting the utility of trajectory-based analysis over static demographic stratification.

This proof-of-concept study demonstrates the feasibility of applying unsupervised temporal modeling to ASD research. While based on simulated data, the findings provide a foundation for applying similar methods to real-world longitudinal datasets. The approach has potential implications for early prognosis, precision intervention, and the development of more developmentally attuned, person-centered frameworks in autism care.




**Introduction**

*1.1 Background*

According to the Centers for Disease Control and Prevention, Autism Spectrum Disorder (ASD) is a complex neurodevelopmental condition affecting approximately 1 in 36 children in the United States. It is marked by wide variability in cognitive abilities, social interaction, communication, and behavior. This phenotypic heterogeneity poses challenges for understanding its etiology, prognosis, and for designing tailored interventions. ASD arises through a combination of genetic and environmental factors, leading to diverse developmental outcomes. While some individuals progress steadily, others experience regression or stagnation, especially in language and adaptive functioning.

Diagnosis typically occurs in early childhood through behavioral assessments such as the Autism Diagnostic Observation Schedule (ADOS) and the Modified Checklist for Autism in Toddlers (M-CHAT). The ADOS evaluates social-communicative behavior in structured settings, while the M-CHAT is a parent-report screener designed to detect early signs of autism risk. Although these tools support early identification, they offer limited predictive insight into long-term developmental trajectories. 

Understanding and mapping these diverse developmental pathways is critical for improving clinical decision-making and advancing personalized support strategies. However, longitudinal characterization of such pathways remains underdeveloped in current research.

*1.2 Limitations of Cross-Sectional Studies in Autism Research*

Cross-sectional research, though practical and widely implemented, captures development at only a single point in time. This design limits the capacity to observe change, thereby constraining our understanding of how cognitive and behavioral profiles evolve over the course of development. For example, two children with comparable language skills at age five may follow divergent developmental paths—one achieving fluent speech by age eight, while the other remains minimally verbal. Such within-person variability is obscured in cross-sectional analyses, which may lead to misinterpretation of developmental trajectories and inaccurate generalizations. Furthermore, cross-sectional designs are susceptible to cohort effects, wherein observed differences between groups may reflect historical, cultural, or contextual influences rather than genuine developmental change.

The limitations of cross-sectional designs have been critically examined in prior work. Gotham et al. (2012), for instance, demonstrated that symptom severity in ASD may appear stable when assessed cross-sectionally, but reveal considerable intra-individual fluctuation when tracked longitudinally. Likewise, Lord et al. (2006) cautioned that static comparisons risk overlooking critical periods of regression, acceleration, or plateauing. These findings collectively underscore the necessity for longitudinal methodologies capable of capturing the temporal complexity and heterogeneity of autistic development.




*1.3 Advantages of Longitudinal Research in Autism Studies*

Longitudinal research designs, which involve repeated measurement of the same individuals over time, are essential for capturing intra-individual developmental change. These methods enable researchers to distinguish between growth, regression, and plateauing—patterns that are particularly salient in the context of Autism Spectrum Disorder (ASD). For example, growth mixture modeling—a statistical technique that identifies unobserved subgroups based on individual trajectories—has been used to uncover distinct symptom pathways that would be obscured in cross-sectional analyses (Gotham et al., 2012; Lord et al., 2015).

Longitudinal data also support the identification of early predictors of later outcomes. For instance, Landa et al. (2013) demonstrated that behavioral indicators observed at 18 months could reliably forecast social communication abilities at age four. This predictive insight is invaluable for early intervention planning and resource allocation.

Furthermore, longitudinal designs facilitate the evaluation of intervention durability. Programs such as the Early Start Denver Model (ESDM) have shown sustained improvements in cognitive functioning and adaptive behavior well beyond the intervention period, as documented in follow-up studies (Dawson et al., 2010; Vivanti et al., 2014). These findings underscore the necessity of tracking outcomes over time to assess long-term efficacy.

Importantly, longitudinal research foregrounds intra-individual variability and reduces reliance on group averages. This is particularly crucial in ASD, where developmental trajectories are highly individualized. Analytical methods such as longitudinal k-means and latent class growth modeling allow researchers to detect meaningful developmental subtypes based on temporal dynamics, rather than relying solely on static measures.

*1.4 Underutilization of longitudinal Methods in ASD research*

Despite methodological advantages, longitudinal study designs remain underutilized in autism research. Several factors contribute to this gap, including the high financial and logistical demands of tracking participants over extended periods, the risk of attrition, and the analytic complexity involved in handling uneven follow-up intervals or evolving assessment tools. Advanced modeling approaches such as latent class growth analysis and functional data clustering also require substantial sample sizes and specialized statistical expertise, which may be prohibitive for many research teams.

Moreover, access to comprehensive, open-source longitudinal datasets remains limited. Although initiatives like the Simons Foundation’s SPARK, the National Database for Autism Research (NDAR), and the Autism Phenome Project (APP) have begun to fill this gap, their data are still underleveraged in empirical work. A bibliometric analysis by Bishop-Fitzpatrick and Kind (2017) noted that fewer than 10% of published autism studies over the past decade employed longitudinal designs, underscoring the dominance of cross-sectional methodologies.

Consequently, most clustering analyses in ASD continue to rely on static, cross-sectional features (Georgiades et al., 2013; Wiggins et al., 2017), potentially overlooking critical within-person patterns of change. This methodological limitation constrains the ability to identify meaningful developmental subtypes.

Addressing this gap presents a significant opportunity for innovation. Expanding the use of longitudinal methods, particularly when paired with unsupervised learning, can unlock deeper insights into the diversity of developmental trajectories and foster more precise, time-sensitive models of care.

*1.5 Rationale for Longitudinal Clustering in ASD*

Clustering algorithms are a class of unsupervised machine learning techniques that identify natural groupings within data based on similarity, without the need for predefined labels. In autism research, clustering has been widely used to detect phenotypic subtypes based on behavioral, cognitive, or neurobiological features. However, these studies have predominantly relied on static measurements—snapshots of functioning at a single time point.

Longitudinal clustering, by contrast, incorporates time-dependent features and groups individuals based on the shape and trajectory of their development. This allows for classification based on dynamic change rather than fixed state. For instance, instead of clustering individuals solely by language ability at age 4, longitudinal approaches consider how language evolves from early childhood into adolescence, thereby identifying developmental subtypes such as "delayed but improving" or "early plateau."

Existing applications of longitudinal models in ASD have largely focused on symptom severity or adaptive functioning trajectories (Gotham et al., 2012; Szatmari et al., 2015), typically using parametric methods such as growth mixture modeling or latent class growth analysis. These approaches, while powerful, rely on predefined assumptions about the shape and number of trajectories—often limiting flexibility.

In contrast, this study applies non-parametric, data-driven longitudinal clustering techniques—methods that do not assume a specific functional form of development—to uncover novel developmental subtypes. This approach is particularly suitable for complex and heterogeneous conditions like ASD, where individual trajectories may follow nonlinear or unexpected paths. By simulating and analyzing synthetic developmental data, this study offers a proof-of-concept for applying such methods in real-world research and clinical contexts.

**Methods**

*2.1 Dataset Description*

To evaluate the feasibility and utility of longitudinal clustering techniques in modeling developmental heterogeneity within Autism Spectrum Disorder (ASD), this study employed a fully synthetic dataset simulating cognitive development from early childhood through adolescence. The simulation parameters were informed by empirical patterns observed in large-scale autism research initiatives, including the National Database for Autism Research (NDAR), the Simons Simplex Collection (SSC), and the SPARK project

The dataset was intentionally designed to reflect demographic diversity and support equity-focused modeling. Each simulated participant was randomly assigned a gender (male, female, or non-binary) and an ethnic identity (White, Black or African American, Hispanic/Latino, Asian, Indigenous, or Multiracial), reflecting the heterogeneity of real-world autism populations. This structure not only enhances ecological validity, but also mitigates ethical concerns related to identifiable patient data and historical biases in ASD research.

The simulated dataset includes the following characteristics:
- Sample size: 200 unique individuals
- Time points: Observations at ages 3, 6, 9, 12, and 15 years
- Primary variable: Z-scored cognitive performance, representing generalized developmental ability. 
- Latent developmental groups(used only for post hoc validation): 
   - Improving: Cognitive scores increase across time
   - Declining: Cognitive scores decrease across time
   - Stable: Cognitive scores remain relatively constant

To simulate natural measurement conditions, Gaussian noise was added at each time point to emulate real-world assessment variability. Although synthetic, the dataset was constructed to mirror several critical properties of clinical longitudinal data, including:
- Within-individual temporal dynamics
- Between-individual heterogeneity
- Natural variation and error in observed scores

By eliminating real-world privacy constraints while preserving methodological integrity, this dataset provides an ethical, flexible foundation for testing unsupervised clustering methods. It enables proof-of-concept exploration of how developmental subtypes may emerge from time-sensitive, data-driven modeling approaches.



*2.2 Data preprocessing*

```python
#Import necessary libraries 
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics import adjusted_rand_score 
from sklearn.preprocessing import StandardScaler
```

```python
# Set seed for reproducibility 
np.random.seed(42)

# Simulation parameters 
n_individuals = 200
ages = [3, 6, 9, 12, 15]
traj_types =['improving', 'declining', 'stable']
data = []

# Generate Simulated Cognitive Trajectories 
for i in range(n_individuals):
    baseline = np.random.normal(loc=70, scale=15)  # Baseline cognitive ability
    traj_type = np.random.choice(traj_types, p=[0.4, 0.3, 0.3])
    
    for j, age in enumerate(ages):
        noise = np.random.normal(loc=0, scale=5)  # Add measurement noise

        # Simulate score based on trajectory type
        if traj_type == 'improving':
            score = baseline + (j * np.random.uniform(1, 3)) + noise
        elif traj_type == 'declining':
            score = baseline - (j * np.random.uniform(1, 3)) + noise
        else:
            score = baseline + noise

        data.append({
            'ID': i,
            'Age': age,
            'Cognitive_Score': round(score, 1),
            'Trajectory_Type': traj_type
        })

# Create DataFrame
df_simulated = pd.DataFrame(data)

# Normalize Cognitive Scores (Z-score)
scaler = StandardScaler()
df_simulated['Cognitive_Score_Normalized'] = scaler.fit_transform(df_simulated[['Cognitive_Score']])

# Simulate Inclusive Demographics
genders = ['Male', 'Female', 'Non-binary']
ethnicities = ['White', 'Black or African American', 'Hispanic/Latino', 'Asian', 'Indigenous', 'Multiracial']

# Assign one gender and one ethnicity per individual
demo_df = pd.DataFrame({
    'ID': range(n_individuals),
    'Gender': np.random.choice(genders, n_individuals),
    'Ethnicity': np.random.choice(ethnicities, n_individuals)
})

# Merge Demographic Info Back Into Main Dataset
df_preprocessed = df_simulated.merge(demo_df, on='ID')

# Convert to Wide Format for Clustering
df_wide = df_preprocessed.pivot(index='ID', columns='Age', values='Cognitive_Score_Normalized')
df_wide.columns = [f'Score_Age_{int(age)}' for age in df_wide.columns]
df_wide.reset_index(inplace=True)

# Add Demographics to Wide Dataset
df_wide = df_wide.merge(demo_df, on='ID')

# Show the cleaned, structured data
df_wide.head()

# Define once: the columns representing cognitive scores at each age
score_cols = ['Score_Age_3', 'Score_Age_6', 'Score_Age_9', 'Score_Age_12', 'Score_Age_15']
```

*2.4 Feature Engineering*

To prepare for clustering, we engineered additional features to capture individual patterns of change across time. While the raw scores at each age provide valuable input, derived features can enhance the model’s ability to detect meaningful subgroups based on developmental dynamics.

The following features were computed for each individual:

- **Slope**: Net change from age 3 to age 15
- **Range**: Difference between max and min cognitive scores
- **Volatility**: Standard deviation across timepoints (developmental variability)
- **First Difference Metrics**: Change between consecutive age scores

These features are designed to summarize both the **direction** and **stability** of cognitive development—two dimensions that are often key in autism trajectories.

```python

# Ensure score_cols is explicitly defined as a list of column names
score_cols = ['Score_Age_3', 'Score_Age_6', 'Score_Age_9', 'Score_Age_12', 'Score_Age_15']

# Copy wide-format data
df_features = df_wide.copy()

# Slope: Age 15 - Age 3
df_features['slope'] = df_features['Score_Age_15'] - df_features['Score_Age_3']

# Range (renamed to avoid conflict with built-in)
df_features['score_range'] = df_features[score_cols].max(axis=1) - df_features[score_cols].min(axis=1)

# Volatility: Standard deviation
df_features['volatility'] = df_features[score_cols].std(axis=1)

# First differences
df_features['diff_3_6'] = df_features['Score_Age_6'] - df_features['Score_Age_3']
df_features['diff_6_9'] = df_features['Score_Age_9'] - df_features['Score_Age_6']
df_features['diff_9_12'] = df_features['Score_Age_12'] - df_features['Score_Age_9']
df_features['diff_12_15'] = df_features['Score_Age_15'] - df_features['Score_Age_12']
```

*2.5 Clustering Setup*

To identify distinct developmental subgroups, we applied unsupervised clustering to the engineered feature set. This approach allows for grouping individuals based on their longitudinal trajectory patterns without using predefined labels.

We start with **K-Means clustering** on a selected feature matrix composed of raw scores and derived features. The optimal number of clusters is determined using the **elbow method** and **silhouette score**.

Clustering was conducted in two stages:
1. **Initial model** using only normalized score trajectories.
2. **Extended model** using additional engineered features.

This section sets the stage for evaluating whether meaningful subgroups emerge that reflect different developmental profiles.

```python
# Select features for clustering
features_to_use = score_cols + ['slope', 'range', 'volatility', 'diff_3_6', 'diff_6_9', 'diff_9_12', 'diff_12_15']
X = df_features[features_to_use]

# Standardize feature matrix
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Find optimal number of clusters using silhouette score
sil_scores = []
cluster_range = range(2, 10)

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    sil = silhouette_score(X_scaled, labels)
    sil_scores.append(sil)

# Plot silhouette scores
plt.figure()
plt.plot(cluster_range, sil_scores, marker='o')
plt.title("Silhouette Score by Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```

The silhouette score quantifies how well each data point fits within its assigned cluster, relative to other clusters. A higher silhouette score indicates better-defined and more separated clusters.

In the plot above, the silhouette score is highest for **k = 2**, suggesting that the data naturally separates best into two distinct clusters under the current feature space. A secondary peak appears at **k = 6**, which may reflect more granular subgroup structure, albeit with slightly reduced cohesion.

While **k = 2** provides the best global structure, additional clusters (e.g., **k = 3–6**) may offer clinically or developmentally meaningful subtypes. For interpretability and comparison with the original simulation (which modeled 3 trajectory types), we will explore both **k = 2** and **k = 3–6** in subsequent analysis.

This flexibility allows us to examine how clustering resolution impacts subtype characterization and whether finer partitions reveal important developmental variation.


*2.7 Clustering Execution and Evaluation*


Based on silhouette analysis and the structure of the simulated dataset, K-Means clustering will be applied to identify latent developmental subtypes. The primary model uses `k = 3` to mirror the labels (improving, declining, stable), but`k = 2` and `k = 6` are examined to explore alternative levels of granularity.

Clustering was applied to a feature matrix including both raw cognitive scores across ages and derived temporal features (e.g., slope, volatility).


```python
# Ground truth labels (from the long-format dataframe)
true_labels = df_preprocessed.drop_duplicates('ID').sort_values('ID')['Trajectory_Type'].values

# Prepare Data
X_cluster = df_features[features_to_use]
X_scaled = scaler.fit_transform(X_cluster)

# Store results
cluster_results = {}

for k in [2, 3, 6]:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    df_features[f'Cluster_k{k}'] = labels
    ari = adjusted_rand_score(true_labels, labels)
    cluster_results[k] = {
        'labels': labels,
        'ARI': ari,
        'model': kmeans
    }
    print(f"Adjusted Rand Index (k={k}): {ari:.3f}")
```
